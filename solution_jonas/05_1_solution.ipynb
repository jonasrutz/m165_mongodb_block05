```python
import pymongo

# Verbindung zur MongoDB herstellen
myclient = pymongo.MongoClient("mongodb://localhost:27017/")
mydb = myclient["mflix"]  # Datenbank "mflix" verwenden
movies_collection = mydb["movies"]  # Collection "movies"

# 1. Geben Sie alle Filme aus, wo die beiden Schauspieler 'Martin' und 'Sinatra' zusammen aufgetreten sind
query1 = {"actors": {"$all": ["/Martin/", "/Sinatra/"]}}
result1 = movies_collection.find(query1, {"title": 1, "_id": 0})
print("Filme mit Martin und Sinatra:")
for movie in result1:
    print(movie)

# 2. Geben Sie alle Filme aus, welche unter anderen Sprachen, sowohl Französisch wie auch Deutsch unterstützen
query2 = {"languages": {"$all": ["French", "German"]}}
result2 = movies_collection.find(query2, {"title": 1, "_id": 0})
print("\nFilme in Französisch und Deutsch:")
for movie in result2:
    print(movie)

# 3. Suche nach Filmen in Italienisch und Französisch
query3 = {"languages": {"$all": ["Italian", "French"]}}
result3 = movies_collection.find(query3, {"title": 1, "_id": 0})
result_count = movies_collection.count_documents(query3)
print("\nFilme in Italienisch und Französisch (Count):", result_count)
for movie in result3:
    print(movie)

# 4. Suche Filme mit einer italienischen Schauspielerin mit "Martin" im Nachnamen
query4 = {"actors": {"$regex": "Martin"}}
result4 = movies_collection.find(query4, {"actors": 1, "_id": 0})
print("\nItalienische Schauspielerin mit 'Martin' im Nachnamen:")
for movie in result4:
    for actor in movie.get("actors", []):
        if "Martin" in actor:
            print(actor)

# 5. Suche nach Filmen mit Jerry Lewis und spezifischen Kriterien
query5 = {"actors": "Jerry Lewis", "year": {"$lt": 1980}, "num_mflix_comments": {"$gt": 25000}}
result5 = movies_collection.find(query5, {"title": 1, "actors": {"$slice": 3}, "_id": 0})
print("\nFilme mit Jerry Lewis und spezifischen Kriterien:")
for movie in result5:
    print(movie)

# 6. Suche nach einem Film mit 8 Oscars und der Geschichte eines Musikers
query6 = {"awards": {"$regex": "8 Oscar"}, "plot": {"$regex": "musician"}}
result6 = movies_collection.find(query6, {"title": 1, "_id": 0})
print("\nFilm mit 8 Oscars über einen Musiker:")
for movie in result6:
    print(movie)

# 7. Suche nach Filmen mit mindestens 99% Tomatometer
query7 = {"tomatoes.viewer.rating": {"$gte": 99}}
result7 = movies_collection.find(query7, {"title": 1, "_id": 0})
result7_count = movies_collection.count_documents(query7)
print("\nFilme mit mindestens 99% Tomatometer:")
print("Anzahl der Filme:", result7_count)
for movie in result7:
    print(movie)
```
